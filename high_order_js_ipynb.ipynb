```python
%%writefile code.js

console.log("Part A")
const data = [5, 12, 3, 20, 11, 9, 30];
let newData = data.filter(x => x > 10).map(x => x*2);
console.log(newData)

console.log("\nPart B")
const numbers = [1, 2, 3, 4, 5, 6];
let num = numbers.filter(x => x%2===0).map(x => x*2).reduce((accumulator, currentValue) => accumulator + currentValue, 0);
console.log(num);

console.log("\nPart C")
const users = [
  { id: 1, name: "Alice", age: 25 },
  { id: 2, name: "Bob", age: 17 },
  { id: 3, name: "Carol", age: 19 }
];
let under18 = users.some(x => x["age"] < 18);
console.log(under18);

let id3 = users.find(x => x["id"] === 3);
console.log(id3);

users.forEach(x => console.log(x["name"]));

console.log("\nPart D");
function repeat(n, action) {
  for (let i=0; i < n; i++) {
    action(i);
  }
}
repeat(3, console.log); // 0, 1, 2
repeat(4, i => console.log("Step " + i));
```
